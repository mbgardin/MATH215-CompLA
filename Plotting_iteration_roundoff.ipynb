**Problem 5**

```python
def g(x):
  return x**4 - 2*x**3 - 17*x**2 + 4*x + 30

def g_prime(x):
  return 4*x**3 - 6*x**2 - 34*x + 4
```

**Problem 6**

```python
def newtons_method(starting_guess,n):
  x_j = starting_guess
  for i in range(n):
    x_j = x_j - g(x_j) / g_prime(x_j)
  return x_j
```

**Problem 7**

```python
def integration(m):
  E = 1 - 1 / np.exp(1)
  for j in range(1, m + 1):
    E = 1 - j * E
  return E
```
